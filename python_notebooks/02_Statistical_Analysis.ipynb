```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_orders.csv")
```

```python
#Mean Revenue
df["Revenue"].mean()
```

```python
#Median Revenue
df["Revenue"].median()
```

```python
#Standard Deviation of Revenue
df["Revenue"].std()
```

```python
#Detect Outliers (IQR Method)
Q1 = df["Revenue"].quantile(0.25)
Q3 = df["Revenue"].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = df[(df["Revenue"] < lower) | (df["Revenue"] > upper)]
len(outliers)
```

```python
#Check Skewness
df["Revenue"].skew() #Negatively skwed distribution, mean < median
```

```python
#Correlation
df["Quantity"].corr(df["Revenue"])
```

Quantity and Revenue have a strong positive correlation. As customers purchase more units, the revenue per order generally increases.
